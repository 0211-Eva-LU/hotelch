```python
from datetime import date, timedelta, datetime

def parse_ymd(s: str) -> date:
    return datetime.strptime(s, "%Y/%m/%d").date()

def datet(checkin: date, checkout: date) -> tuple[str, str, str, str, str, str]:
    # 直接拆，不要用 split
    return (
        str(checkin.year), str(checkin.month), str(checkin.day),
        str(checkout.year), str(checkout.month), str(checkout.day),
    )

def check(start: str, end: str) -> None:
    checkin = parse_ymd(start)
    end_date = parse_ymd(end)

    cnt = 0
    while True:
        checkout = checkin + timedelta(days=1)
        if checkout >= end_date:
            break

        cnt += 1
        cy, cm, cd, oy, om, od = datet(checkin, checkout)
        print(cnt, checkin, checkout, "=>", cy, cm, cd, oy, om, od)

        checkin = checkin + timedelta(days=2)

check("2026/01/11", "2026/01/20")
```

```python
hotel_items = driver.find_elements(By.CSS_SELECTOR, ".hotel-list .list-item")
# HOTEL_NAMES = HOTEL_LIST.find_element(By.CSS_SELECTOR,'hotelName')

hotel_list=[]
cnt=0
for item in hotel_items :
    try:
        name = item.find_element(By.CSS_SELECTOR, ".hotelName").text
        distance = item.find_element(By.CSS_SELECTOR, ".position-desc").text
        price = item.find_element(By.CSS_SELECTOR, ".sale").text
        score = item.find_element(By.CSS_SELECTOR, ".score").text
        cnt+=1
    except Exception as e:
        continue
    hotel_list.append({'title':name,'distance':distance,'price':price,'score':score,'checkin':'2026/05/20','checkout':'2026/05/21'})
    print(cnt,name,distance,price,score)
```

```python
import pandas as pd
import matplotlib
```

```python
hotel_listdf = pd.read_json('hotel_list.json')
```

```python
hotel_listdf["price"] = (
    hotel_listdf["price"]
    .astype(str)
    .str.replace("¥", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(int)   # 若有小數改成 float
)
```

```python
hotel_listdf.describe()
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei", "Microsoft YaHei", "SimHei"]
plt.rcParams["axes.unicode_minus"] = False  # 避免負號變方塊

plt.figure(figsize=(8, 5))
plt.scatter(hotel_listdf["price"], hotel_listdf["score"], alpha=0.7)

# 標註最高價的前 5 個
top5 = hotel_listdf.nlargest(5, "price")
for _, r in top5.iterrows():
    plt.text(r["price"], r["score"], str(r["title"])[:8], fontsize=9)

plt.xlabel("Price")
plt.ylabel("Score")
plt.title("Score vs Price (Top expensive labeled)")
plt.show()
```
